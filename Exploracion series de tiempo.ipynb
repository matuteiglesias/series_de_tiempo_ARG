```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Cuales son las series mas importantes?

import urllib.request

df_list = []
for link in links[:10]:
    df = pd.read_csv(urllib.request.urlopen(link), index_col='indice_tiempo')
    df.index = pd.to_datetime(df.index, format = '%Y-%m-%d')
    df_list += [df]
```
